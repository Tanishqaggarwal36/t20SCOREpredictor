```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Cricket Dataset (1).csv")
```

```python
df.head()
```

```python
df.sort_values(by=['batting_team', 'bowling_team',"city"],inplace=True)
```

```python
df["balls"]=120-df["balls_left"]
df.head()
```

```python
df.sort_values(by=['batting_team', 'bowling_team',"city","balls"],inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 0","pp","balls","Top_Order","Middle_Order","Lower_Order","Tail","Pressure","Aggression_Mode","Death_Overs"],inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns="index",inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df=df.sample(43091)
```

```python
df
```

```python
a=df["city"].value_counts()[df["city"].value_counts()>600].index.to_list()
```

```python
x=df.drop("runs_x",axis=1)
y=df["runs_x"]
```

```python
y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from  sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor,DMatrix
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
import joblib
```

```python
trf=ColumnTransformer([("trf",OneHotEncoder(sparse_output=False,drop="first"),['batting_team','bowling_team','city'])],remainder="passthrough")
```

```python
trf.fit_transform(x)
```

```python
joblib.dump(trf, 'column_transformer.pkl')
```

```python
pipe=Pipeline(steps=[("step1",StandardScaler()),("step2",RandomForestRegressor(n_estimators=1000,max_depth=40,random_state=1))])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test,y_pred)*100        
```

```python
import pickle
pickle.dump(pipe,open("pipe.pkl","wb"))
```
